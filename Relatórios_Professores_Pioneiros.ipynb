```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from pandas_profiling import ProfileReport

base  = pd.read_excel('/content/ResultadoComplexov01.01.xlsx')
# base2 = pd.read_excel('/content/baseAnalisev01.01.xlsx')

base.info()
```

```python
base.drop('SIAPE',                axis = 1, inplace=True)
base.drop('CPF',                  axis = 1, inplace=True)
base.drop('VIGENCIA INICIAL',     axis = 1, inplace=True)
base.drop('VIGENCIA FINAL',       axis = 1, inplace=True)
base.drop('SERVIDOR',             axis = 1, inplace=True)
base.drop('INATIVIDADE',          axis = 1, inplace=True)
base.drop('DATA EXCLUSÃO',        axis = 1, inplace=True)
base.drop('PROCESSO SEI',         axis = 1, inplace=True)
base.drop('DATA DO REQUERIMENTO', axis = 1, inplace=True)
base.drop('INTERESSADO',          axis = 1, inplace=True)
base.drop('DR5',                  axis = 1, inplace=True)
base.drop('COD_NIVELDR',          axis = 1, inplace=True)
base.drop('COD_NIVEL',            axis = 1, inplace=True)
base.drop('INCREMENTO_CNR',       axis = 1, inplace=True)
base.drop('Unnamed: 0',           axis = 1, inplace=True)
base.drop('NIVEL',                axis = 1, inplace=True)
base.drop('NIVEL POR DATA',       axis = 1, inplace=True)
```

```python
base = base.rename(columns={'CARGO':'Cargo Ocupado Pelo Servidor', 
                    'SITUAÇÃO':'Situação no SIAPE', 
                    'OBITO':'Ocorrência de Óbito Após o Requerimento', 
                    'EXCLUSÃO':'Justificativa da Exclusão no SIAPE', 
                    'SITUAÇÃO CADASTRO':'Situação Cadastral do Servidor', 
                    'ESTADO':'Estado', 
                    'NivelEmDR':'Cargo e Nível na Data do Requerimento', 
                    'REGRA':'Regra de Progressão Simplificada', 
                    'CNR':'Classe e Nível do Servidor Calculado Pelo Script', 
                    'RESULTADO':'Resultado da Análise', 
                    'CNRstatus':'Justificativa do Resultado'})

base.info()
```

```python
base.loc[base['Justificativa da Exclusão no SIAPE'] == '00000 -  - ', ['Justificativa da Exclusão no SIAPE']] = 'Cad Ativo'
```

```python
from matplotlib.pyplot import title
from pandas_profiling import ProfileReport

profile = base.profile_report(
    title="Relatório de Análise de Requerimentos de Servidores em Situação: 01 - ATIVO PERMANENTE ou 08 - ATIVO EM OUTRO ORGAO",
    correlations=None,
    samples=None,
    missing_diagrams=None,
    duplicates=None,
    interactions=None,
)

profile.to_file('Relatório de Análise de Requerimentos de Servidores em Situação: 01 - ATIVO PERMANENTE ou 08 - ATIVO EM OUTRO ORGAO.html')
```
